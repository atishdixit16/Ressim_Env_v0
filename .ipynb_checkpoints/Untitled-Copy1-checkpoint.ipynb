```python
import numpy as np
from tqdm import trange
import copy
import matplotlib.pyplot as plt

from ressim_enviroment import resSimEnv_v0, resSimEnv_v1
```

```python
env = resSimEnv_v0(5, n_steps=60, dt=1e-3)
n_action = env.action_space.n
n_episodes = 4
n_steps = 4
cumRewards = []
for i in range(n_episodes):
    state = env.reset()
    cumR = 0.0
    for j in range(n_steps):
        env.render(i,j)
        action = np.random.randint(low=0, high=n_action)
        state, reward, done, info = env.step(action)
        cumR += reward
    cumRewards.append(cumR)
    print('Episode: '+str(i+1)+', reward:'+str(round(cumR,1)))
```

```python
env = resSimEnv_v1(11, n_steps=60, dt=1e-3)
n_action = env.action_space.n
n_episodes = 5
n_steps = 4
cumRewards = []
for i in range(n_episodes):
    state = env.reset()
    cumR = 0.0
    for j in range(n_steps):
        env.render(i,j)
        action = np.random.randint(low=0, high=n_action)
        state, reward, done, info = env.step(action)
        cumR += reward
    cumRewards.append(cumR)
    print('Episode: '+str(i+1)+', reward:'+str(round(cumR,1)))
    
```
